```python
import numpy as np

# Создаем два массива размерности (10, 4) со случайными целыми числами от -5 до 5
ar1 = np.random.randint(-5, 6, size=(10, 4))
ar2 = np.random.randint(-5, 6, size=(10, 4))

# Удваиваем значения в ar1, где они больше соответствующих значений в ar2, остальные делаем 0
result = np.where(ar1 > ar2, ar1 * 2, 0)

print("Массив ar1:")
print(ar1)
print("\nМассив ar2:")
print(ar2)
print("\nРезультат:")
print(result)
```

```python
import numpy as np

# Создаем массив размерности (25, 4) со случайными целыми числами от 0 до 15
ar1 = np.random.randint(0, 16, size=(25, 4))

# Находим максимальные значения в каждой строке
max_in_rows = np.max(ar1, axis=1, keepdims=True)

# Создаем массив, где True - если элемент максимальный в своей строке
is_max = (ar1 == max_in_rows)

# Считаем количество максимальных элементов в каждом столбце
max_counts_per_col = np.sum(is_max, axis=0)

# Находим столбцы, где не менее 5 максимальных значений
cols_with_5max = np.where(max_counts_per_col >= 5)[0]
print(f"Столбцы, где не менее 5 раз встречается максимальное значение строки: {cols_with_5max}")

# Находим столбец с наибольшим количеством максимумов
col_with_most_max = np.argmax(max_counts_per_col)

# Заменяем максимумы по строке на -1 в этом столбце
ar1[(is_max[:, col_with_most_max]), col_with_most_max] = -1

print("\nМассив после замены:")
print(ar1)
```

```python
import numpy as np

# Создаем массив 30x4 со случайными целыми числами от 0 до 100
vectors = np.random.randint(0, 101, size=(30, 4))

# Вычисляем евклидову норму (длину) каждого вектора
norms = np.linalg.norm(vectors, axis=1)

# Находим индексы 5 векторов с наибольшей длиной
top5_indices = np.argsort(norms)[-5:]

# Получаем сами векторы с наибольшей длиной
top5_vectors = vectors[top5_indices]

print("Исходный массив векторов:")
print(vectors)
print("\n5 векторов с наибольшей длиной:")
print(top5_vectors)
```

```python
import numpy as np

def one_hot_encoding(arr):
    # Определяем количество уникальных классов (начинаем с 0)
    num_classes = np.max(arr) + 1

    # Создаем матрицу one-hot encoding
    encoding = np.zeros((len(arr), num_classes), dtype=int)

    # Заполняем единицами соответствующие позиции
    encoding[np.arange(len(arr)), arr] = 1

    return encoding

# Тестирование на случайном массиве
random_array = np.random.randint(0, 5, size=10)  # Числа от 0 до 4, длина 10
print("Исходный массив:")
print(random_array)
print("\nOne-hot encoding:")
print(one_hot_encoding(random_array))

# Тест на примере из задания
test_array = np.array([2, 3, 2, 2, 2, 1])
print("\nТестовый массив:")
print(test_array)
print("\nOne-hot encoding для тестового массива:")
print(one_hot_encoding(test_array))
```

```python
import numpy as np

# Создаем массив 10x4 со случайными целыми числами от 0 до 50
arr = np.random.randint(0, 51, size=(10, 4))

# Находим уникальные элементы и их частоты
unique, counts = np.unique(arr, return_counts=True)

# Находим максимальную частоту
max_count = np.max(counts)

# Находим все числа с максимальной частотой
most_frequent_numbers = unique[counts == max_count]

print("Исходный массив:")
print(arr)
print("\nСамое частое число(а):", most_frequent_numbers)
print("Количество вхождений:", max_count)

# Создаем маску для всех позиций, где встречаются самые частые числа
mask = np.isin(arr, most_frequent_numbers)

# Считаем количество самых частых чисел в каждой строке
counts_per_row = np.sum(mask, axis=1)

# Находим индексы 3 строк с наибольшим количеством частых чисел
top3_rows = np.argsort(counts_per_row)[-3:]

# Получаем итоговый массив 3x4
result = arr[top3_rows]

print("\nСтроки с наибольшим количеством самых частых чисел:")
print(result)
```
